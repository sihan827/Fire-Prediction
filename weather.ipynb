```python
import pandas as pd
import numpy as np
import requests 
import json
```

```python
df_T = pd.read_csv('./data/temperature_humidity.csv')
df_W = pd.read_csv('./data/wind_speed_rain.csv')
```

```python
# 일시, 평균기온(°C), 최저기온(°C), 최고기온(°C), 평균 상대습도\n(%)
df_T = df_T.loc[:, ['일시', '평균기온(°C)', '최저기온(°C)', '최고기온(°C)', '평균 상대습도\n(%)']]
df_T = df_T.fillna(0)
df_T.head()
```

```python
# 일강수량(mm), 평균 풍속(m/s)
df_W = df_W.loc[:, ['일강수량(mm)', '평균 풍속(m/s)']]
df_W = df_W.fillna(0)
df_W.head()
```

```python
df = pd.concat([df_T, df_W], axis=1)
df.head()
```

```python
df.to_csv('./data/preprocessed_weather.csv', encoding='utf-8-sig', index=False)
```
